#Topic: Introduction to OOP

**1. Evolution:**

* Procedural Programming (The Past): Focused on functions and logic steps. Data was global and separated from logic. Hard to maintain in large projects.

* OOP (The Solution): Bundles data and behavior together, mimicking real-world entities.

**Why did OOP come into the picture?**

OOP was invented to solve the "messy data" problem.

Instead of keeping data and functions separate, computer scientists realized: "Why not group the data and the functions that use that data into a single container?"

This container is what we call an Object. This shift made code safer, more organized, and easier to scale.

**What is OOP?**

**Definition:** OOP (Object-Oriented Programming) is a way of structuring programs so that data (attributes) and behavior (methods) are bundled together into objects.


*   Python is an object-oriented language, meaning almost everything is an object.
*   Object Oriented Programming is a fundamental concept in Python, empowering developers to build modular, maintainable and scalable applications.


**Key Features of OOP in Python:**

*   Organizes code into classes and objects
*   Supports encapsulation to group data and methods together
*   Enables inheritance for reusability and hierarchy
*   Allows polymorphism for flexible method implementation
*  Improves modularity, scalability, and maintainability

#**Abstraction**

1. Definition: Abstraction is the concept of hiding the complex implementation details of a system and exposing only the necessary and relevant features to the user. It focuses on the "what" rather than the "how".

2. Purpose & Benefits:

    * Reduces Complexity: Allows users/programmers to interact with complex systems simply.

    * Increases Maintainability: You can change the internal messy code without affecting the code that uses the object.

    * Security: Helps hide critical data from direct external access.

3. Implementation in Python:

    * Convention: Python primarily uses the single underscore naming convention (e.g., _internal_method) to signal that a method or attribute is intended for internal use within the class and should be abstracted away from the end-user.

    * Abstract Base Classes (ABCs): (Advanced) Python's abc module allows for defining formal interfaces that enforce structure in subclasses

Example of Automatic Coffee Machine

```python
class CoffeeMachine:
    def __init__(self):
        self.water_level = 100 # Internal data

    # --- Internal complex steps (Hidden details) ---
    # Note the underscore "_" at the start. This is a hint these are internal.
    def _boil_water(self):
        print("...Internal process: Boiling water...")

    def _grind_beans(self):
        print("...Internal process: Grinding beans...")

    def _pump_water(self):
        print("...Internal process: Pumping water through grounds...")

    # --- The Public Interface (What the user sees) ---
    def make_coffee(self):
        print("Starting coffee procedure!")
        # The complex steps happen internally
        self._boil_water()
        self._grind_beans()
        self._pump_water()
        print("Your coffee is ready! Enjoy.")
```

```python
my_machine = CoffeeMachine()
my_machine.make_coffee()
```

##Encapsulation

1. Definition: Encapsulation is the practice of bundling data and methods within a single unit (Class) and restricting direct access to some of an object's internal components.

2. Goals:

    * Protection: Prevents accidental modification of data from outside the class.

    * Control: Ensures data is modified only through well-defined methods (Getters/Setters) that can include validation logic.

3. Python Implementation:

    * Public Members (var): Accessible from anywhere.

    * Protected Members (_var): Indication that the member should not be accessed outside the class or its subclasses (Convention only).

    * Private Members (__var): Strongly hidden. Python performs "name mangling" (changing the name internally) to make it difficult to access from outside.

4. Getters and Setters: Methods used to retrieve (get) or modify (set) the value of a private variable, allowing for validation rules (e.g., ensuring a bank balance never goes below zero).

```python
class BankAccount:
    def __init__(self, owner, initial_deposit):
        self.owner = owner
        # Double underscore makes it PRIVATE.
        # It cannot be accessed directly from outside.
        self.__balance = initial_deposit

    # Method to ADD money (Setter logic)
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}")
        else:
            print("Invalid deposit amount")
    def withdraw(self,amount):
      if amount > 0:
        self.__balance -= amount
        print(f"Debited in your account ${amount} and the remaining balance is {self.__balance}")

    # Method to CHECK balance (Getter logic)
    def check_balance(self):
        # We can verify user identity here if we wanted to
        return self.__balance

# --- Usage ---
my_account = BankAccount("John", 0)
my_account.deposit(800)
```

```python
my_account.withdraw(200)
```

## Inheritance
1. Definition: The capability of a class to derive or inherit the properties (attributes) and methods from another class.

2. Terminology:

    * Super Class (Parent): The class being inherited from.

    * Sub Class (Child): The class that inherits. It can access parent features and extend them.

3. Key Benefits:

    * Reusability: Write common logic once in the parent class and reuse it across multiple child classes.

    * Extensibility: Add new features to an existing class without modifying the original code.

4. Types supported in Python:

    * Single: One to One.

    * Multilevel: Grandparent -> Parent -> Child.

    * Multiple: One Child inherits from multiple Parents (e.g., class Child(Parent1, Parent2)).

    * Hierarchical: Multiple Children inherit from one Parent.

5. super(): A built-in function that returns a proxy object (temporary reference) of the parent class, allowing you to call methods defined in the superclass (commonly used in __init__).

A. Single Inheritance
One Child inherits from one Parent.

```python
# Parent Class
class Animal:
    def speak(self):
        print("Animal Speaking...")

# Child Class inherits from Animal
class Dog(Animal):
    def bark(self):
        print("Woof!")

d = Dog()
d.speak()  # Inherited from Animal (Output: Animal Speaking...)
d.bark()   # Defined in Dog (Output: Woof!)
```

B. Multilevel Inheritance
A chain of inheritance. Child inherits from Parent, who inherited from Grandparent.

```python
class Grandparent:
    def farm(self):
        print("I have land")

class Parent(Grandparent):
    def house(self):
        print("I have a house")

class Child(Parent):
    def car(self):
        print("I have a car")

c = Child()
c.farm()   # Inherited from Grandparent
c.house()  # Inherited from Parent
c.car()    # Own feature
```

C. Multiple Inheritance
A Child inherits from two or more Parents at the same time. (Note: Java doesn't support this, but Python does!)

```python
class Father:
    def gardening(self):
        print("I enjoy gardening")

class Mother:
    def cooking(self):
        print("I enjoy cooking")

# Child inherits from BOTH
class Child(Father, Mother):
    def sports(self):
        print("I enjoy sports")

c = Child()
c.gardening() # From Father
c.cooking()   # From Mother
c.sports()    # Own
```

4. The super() Function
Sometimes, you want to use the Parent's method but add something extra to it. You use super().

Example: A Manager is an Employee, but gets a bonus added to their salary.

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class Manager(Employee):
    def __init__(self, name, salary, bonus):
        # Use super() to let the Parent handle the standard stuff
        super().__init__(name, salary)
        self.bonus = bonus  # Handle the extra stuff manually

m = Manager("Rahul", 50000, 10000)
print(m.salary) # 50000
print(m.bonus)  # 10000
```

##Polymorphism

1. Definition: The ability of different objects to respond to the same function call (method name) in different ways. It allows a single interface to control a variety of underlying forms.

2. Method Overriding (Runtime Polymorphism):

    * Occurs when a Child class provides a specific implementation of a method that is already defined in its Parent class.

    * The method name and parameters remain the same, but the logic inside changes.

3. Benefits:

    * Simplicity: The programmer only needs to remember one method name (e.g., save()) rather than distinct names for every type (e.g., save_pdf(), save_doc()).

    * Extensibility: You can introduce new classes (e.g., a new Payment Method) without breaking existing code that uses the interface.

4. Duck Typing (Python Specific): Python's dynamic nature allows polymorphism without strict inheritance. If an object implements the expected method, Python will accept it, regardless of its class hierarchy.

```python
class Animal:
    def speak(self):
        print("Some generic animal sound")

class Dog(Animal):
    # Same method name, different behavior (Overriding)
    def speak(self):
        print("Woof!")

class Cat(Animal):
    # Same method name, different behavior (Overriding)
    def speak(self):
        print("Meow!")

# --- The Power of Polymorphism ---

# Imagine we have a list of random animals
zoo = [Dog(), Cat(), Dog(), Animal()]

for animal in zoo:
    # We call the SAME method 'speak()' for every object.
    # Python automatically knows which version to use.
    animal.speak()
```

```python
# 1. Parent Class (The Blueprint)
class PaymentMethod:
    def pay(self, amount):
        pass # Defined in children

# 2. Child Classes (The Implementations)
class CreditCard(PaymentMethod):
    def pay(self, amount):
        print(f"Paying ${amount} using Credit Card Logic (Stripe)")

class PayPal(PaymentMethod):
    def pay(self, amount):
        print(f"Paying ${amount} using PayPal Logic (API Login)")

# 3. The "Polymorphic" Function
# This function doesn't care if it's a card or paypal.
# It just knows it can call .pay()
def process_customer_order(payment_obj, cost):
    payment_obj.pay(cost)

# --- Usage ---
card = CreditCard()
paypal = PayPal()

# The same function handles both objects seamlessly!
process_customer_order(card, 100)   # Output: Paying $100 using Credit Card Logic...
process_customer_order(paypal, 50)  # Output: Paying $50 using PayPal Logic...
```

```python
class Car:
    def move(self):
        print("Drive")

class Boat:
    def move(self):
        print("Sail")

# These two are unrelated, but because they both have "move()",
# we can treat them polymorphically.
for vehicle in [Car(), Boat()]:
    vehicle.move()
```
