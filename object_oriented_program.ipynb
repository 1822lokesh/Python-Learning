#Topic: Introduction to OOP

**1. Evolution:**

* Procedural Programming (The Past): Focused on functions and logic steps. Data was global and separated from logic. Hard to maintain in large projects.

* OOP (The Solution): Bundles data and behavior together, mimicking real-world entities.

**Why did OOP come into the picture?**

OOP was invented to solve the "messy data" problem.

Instead of keeping data and functions separate, computer scientists realized: "Why not group the data and the functions that use that data into a single container?"

This container is what we call an Object. This shift made code safer, more organized, and easier to scale.

**What is OOP?**

**Definition:** OOP (Object-Oriented Programming) is a way of structuring programs so that data (attributes) and behavior (methods) are bundled together into objects.


*   Python is an object-oriented language, meaning almost everything is an object.
*   Object Oriented Programming is a fundamental concept in Python, empowering developers to build modular, maintainable and scalable applications.


**Key Features of OOP in Python:**

*   Organizes code into classes and objects
*   Supports encapsulation to group data and methods together
*   Enables inheritance for reusability and hierarchy
*   Allows polymorphism for flexible method implementation
*  Improves modularity, scalability, and maintainability





**Classes and Objects**

* Class: The blueprint or template. It defines what data (attributes) and actions (methods) an object will have.

* Object: An instance of a class. It's the actual "thing" created from the blueprint.

```python
class Employee:
  def __init__(self,name,age,role,salary):
    self.name=name
    self.age=age
    self.role=role
    self.salary=salary
    print("These are the employee detalis ")

  def info(self):
    print(f"{self.name} is {self.age} years old and he working as {self.role} and he is earning monthly {self.salary}")
```

```python
em1=Employee("Arun",30,"Software Developer",45000)
em1.info()
```

**Key Python Keywords:**

* class Name:: Defines a new class. (Standard practice is to Capitalize class names, e.g., MyClass).

* __init__(self): The initializer method. It is automatically called when a new object is created to assign values to object properties.

* self: A reference to the current instance of the class. It allows access to the attributes and methods of the specific object being interacted with.
