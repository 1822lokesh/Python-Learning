Example of ATM Machine

```python
class ATM:
    def __init__(self):
        # Using private variables (__) to hide sensitive data
        self.__pin = ""
        self.__balance = 0
        self.menu()

    def menu(self):
        while True:
            user_input = input("""
            Hello, how would you like to proceed?
            [1] Create Pin
            [2] Change Pin
            [3] Check Balance
            [4] Deposit
            [5] Withdraw
            [6] Exit
            Enter Option: """)

            if user_input == '1':
                self.create_pin()
            elif user_input == '2':
                self.change_pin()
            elif user_input == '3':
                self.check_balance()
            elif user_input == '4':
                self.deposit()
            elif user_input == '5':
                self.withdraw()
            elif user_input == '6':
                print("Thank you for using our ATM. Goodbye!")
                break
            else:
                print("Invalid option. Please try again.")

    # Helper method to avoid repeating code (DRY Principle)
    def _verify_pin(self):
        if self.__pin == "":
            print("Please create a pin first (option 1).")
            return False

        entered_pin = input("Enter your PIN: ")
        if entered_pin == self.__pin:
            return True
        else:
            print("Wrong PIN.")
            return False

    def create_pin(self):
        if self.__pin != "":
            print("PIN already exists. Use 'Change Pin' option.")
            return

        user_pin = input("Create your PIN: ")
        self.__pin = user_pin
        print("PIN created successfully!")

    def change_pin(self):
        if self._verify_pin():
            new_pin = input("Enter new PIN: ")
            self.__pin = new_pin
            print("PIN changed successfully!")

    def check_balance(self):
        if self._verify_pin():
            print(f"Current Balance: ${self.__balance}")

    def deposit(self):
        if self._verify_pin():
            try:
                amount = int(input("Enter amount to deposit: "))
                if amount > 0:
                    self.__balance += amount
                    print(f"Deposited ${amount}. New Balance: ${self.__balance}")
                else:
                    print("Amount must be positive.")
            except ValueError:
                print("Invalid input. Please enter a number.")

    def withdraw(self):
        if self._verify_pin():
            try:
                amount = int(input("Enter amount to withdraw: "))
                if amount <= 0:
                    print("Amount must be positive.")
                elif amount > self.__balance:
                    print("Insufficient funds!")
                else:
                    self.__balance -= amount
                    print(f"Withdrawn ${amount}. New Balance: ${self.__balance}")
            except ValueError:
                print("Invalid input. Please enter a number.")
```
