#Python Errors and Exception Handling


##1. Types of Errors

There are two main types of errors in Python:

**Syntax Errors and Exceptions.**


**A. Syntax Errors (Parsing Errors)**

Occur when the code violates the grammatical rules of Python. The program will not start until these are fixed.

* Cause: Missing colons, mismatched parentheses, wrong indentation.

* Example:

```python
# Error: Missing colon
if 5 > 3
    print("Hello")
```

**B. Exceptions (Runtime Errors)**

Occur during execution even if the syntax is correct. The program crashes unless handled.

* Cause: Division by zero, accessing a non-existent file, invalid index.

* Example:

```python
# Error: Division by zero
print(10 / 0)
```

**2. Exception Handling (try ... except)**

Python uses try and except blocks to handle exceptions so the program doesn't crash effectively.

The Structure:

* try: Run this code.

* except: If an error happens in the try block, run this code.

* else: If no error happens, run this.

* finally: Always run this code, regardless of errors (used for cleanup).

Syntax Structure

```python
try:
    # Code that might cause an error
    pass
except ExceptionName:
    # Code that runs if that specific error occurs
    pass
else:
    # Code that runs if NO error occurs
    pass
finally:
    # Code that runs ALWAYS (cleanup)
    pass
```

**Example: Complete Workflow**

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
    print(f"Result is: {result}")

except ZeroDivisionError:
    print("Error: You cannot divide by zero.")

else:
    print("Calculation successful.")

finally:
    print("Execution complete.")
```

**3. Raising Exceptions (raise)**

You can force an exception to occur using the raise keyword. This is useful for enforcing rules in your code.

Example

```python
age = -5

if age < 0:
    raise ValueError("Age cannot be negative.")
```
